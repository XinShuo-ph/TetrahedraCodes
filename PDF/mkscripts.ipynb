```python
# this is very fast, probably no need for submitting to SLURM
basename = 'getfinalPDF_Gadget_CIC'
scriptdir = basename + 'Jobs/'

# try an example script

Rsmooth= 5
simno = 0
snapno  = 4

hours = 24
GBs = 250

jobname=basename + "_R%d_sim%d_snap%d"%(Rsmooth,simno,snapno)
jobtime="%02d:00:00"%hours
jobmem ="%dG"%GBs
jobout =scriptdir+basename + "_R%d_sim%d_snap%d.out"%(Rsmooth,simno,snapno)
joberr =scriptdir+basename + "_R%d_sim%d_snap%d.err"%(Rsmooth,simno,snapno)
module = "\nmodule load openmpi\nmodule load gsl\nmodule load openblas\n\n"
jobcmd =module+ "python3 "+basename+".py %d %d %d"%(Rsmooth,simno,snapno)
jobcmd = jobcmd + " > " + jobout
myscript="#!/bin/bash \n#SBATCH --job-name=" + jobname + " \n#SBATCH --time=" + jobtime + " \n#SBATCH --ntasks=2 \n#SBATCH --cpus-per-task=32 \n#SBATCH --mem=" + jobmem + " \n#SBATCH -p hns,kipac \n#SBATCH --output=" + jobout + " \n#SBATCH --error=" + joberr + " \n\n" + jobcmd

print(myscript)
```
