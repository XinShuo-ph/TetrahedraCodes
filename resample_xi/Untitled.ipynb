```python
import numpy as np
from colossus.cosmology import cosmology
import readgadget
from tetrahedrafunc import *

QuijoteFid = {'H0': 67.11,
 'Ob0': 0.049, 
 'Om0': 0.3175,
 'flat': True,
 'ns': 0.9624,
 'sigma8': 0.834}
mycosmos =  cosmology.setCosmology("QuijoteFid", **QuijoteFid)
```

```python
z = np.linspace(0,50,10000)
scales = mycosmos.growthFactor(z)
```

```python
uniformlogscales100 = np.linspace(max(np.log(scales)),min(np.log(scales)),100)
```

```python
uniformlogscales100
```

```python
uniformscalezs = []

cur = 0
for idx in range(len(scales)):
    if np.log(scales[idx]) <= uniformlogscales100[cur]:
        uniformscalezs.append(z[idx])
        cur +=1

```

```python
uniformscalezs
```

```python
len(uniformscalezs)
```

```python
uniformscalezs = np.array(uniformscalezs)
```

```python
np.log(mycosmos.growthFactor(uniformscalezs))[1::] - np.log(mycosmos.growthFactor(uniformscalezs))[0:-1]
```

```python
ts = 1.0/(1+uniformscalezs)
```

```python
with open("uniscale_times.dat","w") as f:
    for t in ts[::-1]:
        f.write("%.6f\n"%t)
```
